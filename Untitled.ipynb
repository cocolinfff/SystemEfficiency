```python
from CoolProp.CoolProp import PropsSI
def SI(char,T=-1,P=-1,x=-1,h=-1):
    if (T==-1)+(P==-1)+(x==-1)+(h==-1)!=2:
        print('not enough paras')
        return
    if T==-1 and P==-1:
        return PropsSI(char,'Q',x,'H',h,'Water')
    elif T==-1 and x==-1:
        return PropsSI(char,'P',P,'H',h,'Water')
    elif T==-1 and h==-1:
        return PropsSI(char,'P',P,'Q',x,'Water')
    elif P==-1 and x==-1:
        return PropsSI(char,'T',T,'H',h,'Water')
    elif P==-1 and h==-1:
        return PropsSI(char,'T',T,'Q',x,'Water')
    elif x==-1 and h==-1:
        return PropsSI(char,'T',T,'P',P,'Water')
```

```python
Mpa=1e6 #兆帕斯卡
class node: # node节点是用来已知某点两个参数自动求其余参数的类
    def __init__(self,name,H=-1,T=-1,x=-1,P=-1):
        self.name=name
        self.H=H
        self.T=T
        self.x=x
        self.P=P
        self.S=-1
        
    def display(self):
        print(self.name+':'+'\n\tTemp='+str(self.T)+' K\n\tH='+'%.2f'%(self.H/1e3)+' kJ/kg\n\tPressure='+str(self.P/Mpa)+' Mpa')
        
    def calc(self):
        if self.T==-1 and self.P==-1:
            self.T=PropsSI('T','Q',self.x,'H',self.H,'CarbonDioxide')
            self.P=PropsSI('P','Q',self.x,'H',self.H,'CarbonDioxide')
        elif self.T==-1 and self.x==-1:
            self.T=PropsSI('T','P',self.P,'H',self.H,'CarbonDioxide')
            self.x=PropsSI('Q','P',self.P,'H',self.H,'CarbonDioxide')
        elif self.T==-1 and self.H==-1:
            self.T=PropsSI('T','P',self.P,'Q',self.x,'CarbonDioxide')
            self.H=PropsSI('H','P',self.P,'Q',self.x,'CarbonDioxide')
        elif self.P==-1 and self.x==-1:
            self.P=PropsSI('P','T',self.T,'H',self.H,'CarbonDioxide')
            self.x=PropsSI('Q','T',self.T,'H',self.H,'CarbonDioxide')
        elif self.P==-1 and self.H==-1:
            self.P=PropsSI('P','T',self.T,'Q',self.x,'CarbonDioxide')
            self.H=PropsSI('H','T',self.T,'Q',self.x,'CarbonDioxide')
        elif self.x==-1 and self.H==-1:
            self.Q=PropsSI('Q','T',self.T,'P',self.P,'CarbonDioxide')
            self.H=PropsSI('H','T',self.T,'P',self.P,'CarbonDioxide')
        self.S=PropsSI('S','T',self.T,'P',self.P,'CarbonDioxide')/1000
            

    
```

# 堆芯进口温度的影响
* 已知换热器出口温度为872K   
* 不考虑回热器温度限制条件的在堆芯入口温度 𝑇10 在设置为 863◦C，868◦C，822◦C，852◦C 下，堆芯出口温度 𝑇1 为 973◦C下绘制的若干效率-抽气比曲线。
* 观察可以发现，在这四种边界条件下循环效率与抽气比为负相关，当抽气比趋于 0 时效率达到最大值，而且堆芯入口温度越高整体效率越高，当堆芯入口温度达到一定程度时效率可以在设备非理想状态下达到 35%的循环效率。
* 需要注意的是，此处最低温度点取值为55度，而理论最低温度为31度（超临界二氧化碳的临界温度）
* 可以发现，当入口温度越高，所能取到的效率越高。但实际上入口温度会影响分流比的取值
